```python
import pandas as pd
import graphviz
import numpy as np
from sklearn import datasets
FEATURE_NAMES = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data, columns = FEATURE_NAMES)
y = iris.target
print(X)
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=3) # if we remove max_depth, will get whole tree
model.fit(X,y)
```

```python
from sklearn.tree import export_graphviz
#import graphviz
export_graphviz(model, 'flower.dot', feature_names = FEATURE_NAMES)
```
